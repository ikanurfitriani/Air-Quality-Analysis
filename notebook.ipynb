### Cleaning Data
#### Mengganti nilai hilang pada kolom numerik dengan mean dan kolom kategori dengan mode

```python
numeric_columns = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']
categorical_columns = ['wd']

df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])

print(df.isnull().sum())
```

#### Menghapus outlier

```python
def remove_outliers_iqr(df, columns):
    df_cleaned = df.copy()
    for col in columns:
        Q1 = df_cleaned[col].quantile(0.25)
        Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df_cleaned = df_cleaned.loc[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
    
    return df_cleaned

numeric_columns = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'WSPM']

df_cleaned = remove_outliers_iqr(df, numeric_columns)
```

#### Melihat sebaran data dengan boxplot setelah penghapusan outlier

```python
columns_to_plot = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'TEMP', 'PRES', 'DEWP', 'WSPM']

fig, axes = plt.subplots(5, 2, figsize=(14, 24))

sns.set_theme(style="whitegrid", palette="coolwarm")
axes = axes.flatten()

for i, column in enumerate(columns_to_plot):
    sns.boxplot(
        y=df_cleaned[column], 
        ax=axes[i], 
        color="steelblue",
        linewidth=1.5,  
        fliersize=5, 
        boxprops=dict(edgecolor="navy", linewidth=1.5),  
        whiskerprops=dict(color="darkred", linewidth=1.2),  
        capprops=dict(color="darkred", linewidth=1.2)  
    )
    axes[i].set_title(f"Boxplot of {column}", fontsize=12, fontweight="bold")  
    axes[i].set_ylabel(column, fontsize=10)  
    axes[i].tick_params(axis='both', which='major', labelsize=9)

for j in range(len(columns_to_plot), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout(h_pad=2)  
plt.suptitle("Boxplots of Air Quality Data", fontsize=16, fontweight="bold", color="darkblue")  
plt.subplots_adjust(top=0.94)  
plt.show()
```

### Pertanyaan 2: Bagaimana tren polutan partikulat (PM2.5 dan PM10) di pagi hari dibandingkan dengan sore hari di setiap musim?

```python
def assign_time_of_day(hour):
    if 5 <= hour <= 11:  
        return 'Pagi'
    elif 12 <= hour <= 18:  
        return 'Sore'
    else:
        return 'Malam'

df_cleaned['time_of_day'] = df_cleaned['hour'].apply(assign_time_of_day)

def assign_season(month):
    if month in [12, 1, 2]:
        return 'Dingin'
    elif month in [3, 4, 5]:
        return 'Semi'
    elif month in [6, 7, 8]:
        return 'Panas'
    elif month in [9, 10, 11]:
        return 'Gugur'

df_cleaned['season'] = df_cleaned['month'].apply(assign_season)

seasonal_trends = (
    df_cleaned[df_cleaned['time_of_day'].isin(['Pagi', 'Sore'])]
    .groupby(['season', 'time_of_day'])[['PM2.5', 'PM10']]
    .mean()
    .reset_index()
)

sns.set_theme(style="whitegrid", palette="pastel")

plt.figure(figsize=(12, 8))
sns.barplot(
    data=seasonal_trends,
    x='season', 
    y='PM2.5', 
    hue='time_of_day', 
    errorbar=None, 
    palette="coolwarm",
    edgecolor='black'
)
plt.title('Rata-rata PM2.5 di Pagi dan Sore Hari untuk Setiap Musim', fontsize=16, fontweight='bold')
plt.xlabel('Musim', fontsize=12)
plt.ylabel('Konsentrasi Rata-rata PM2.5 (µg/m³)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Waktu', fontsize=10, title_fontsize=11)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
sns.barplot(
    data=seasonal_trends,
    x='season', 
    y='PM10', 
    hue='time_of_day', 
    errorbar=None, 
    palette="coolwarm",
    edgecolor='black'
)
plt.title('Rata-rata PM10 di Pagi dan Sore Hari untuk Setiap Musim', fontsize=16, fontweight='bold')
plt.xlabel('Musim', fontsize=12)
plt.ylabel('Konsentrasi Rata-rata PM10 (µg/m³)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title='Waktu', fontsize=10, title_fontsize=11)
plt.tight_layout()
plt.show()
```

**Insight:**
<br>
Berdasarkan visualisasi barplot di atas konsentrasi rata-rata PM10 paling rendah di musim dingin pagi dan paling tinggi di musim semi pagi sedangkan PM2.5 paling rendah di musim dingin pagi dan paling tinggi di musim panas pagi. PM10 lebih rendah di musim dingin pagi karena curah hujan atau salju dan suhu yang lebih rendah membantu menurunkan konsentrasi debu dan partikel besar lainnya serta PM10 lebih tinggi di musim semi pagi karena angin yang kuat dan aktivitas pertanian yang lebih banyak menyebabkan peningkatan debu di udara. PM2.5 lebih rendah di musim dingin pagi karena aktivitas manusia terbatas dan emisi polutan yang lebih rendah serta PM2.5 lebih tinggi di musim panas pagi karena suhu tinggi meningkatkan pembakaran bahan bakar dan aktivitas manusia, menghasilkan lebih banyak emisi dari kendaraan dan industri.

### Pertanyaan 3: Apakah ada tren perubahan polutan gas (SO2, NO2, CO, O3) selama periode 2013 hingga 2017?

```python
trend_gases = (
    df_cleaned.groupby(['year', 'month'])[['SO2', 'NO2', 'CO', 'O3']]
    .mean()
    .reset_index()
)

trend_gases['year_month'] = pd.to_datetime(trend_gases[['year', 'month']].assign(day=1))

sns.set_theme(style="whitegrid", palette="muted")

plt.figure(figsize=(16, 10))

sns.lineplot(data=trend_gases, x='year_month', y='SO2', label='SO2', color='darkblue', linewidth=2.5)
sns.lineplot(data=trend_gases, x='year_month', y='NO2', label='NO2', color='green', linewidth=2.5)
sns.lineplot(data=trend_gases, x='year_month', y='CO', label='CO', color='red', linewidth=2.5)
sns.lineplot(data=trend_gases, x='year_month', y='O3', label='O3', color='purple', linewidth=2.5)

plt.title('Trend Perubahan Polutan Gas (SO2, NO2, CO, O3) (2013-2017)', fontsize=18, fontweight='bold', color='navy')
plt.xlabel('Periode Waktu', fontsize=14)
plt.ylabel('Konsentrasi Rata-rata (\u03bcg/m³)', fontsize=14)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12)

plt.grid(visible=True, which='both', linestyle='--', linewidth=0.5)
plt.legend(title='Polutan Gas', fontsize=12, title_fontsize=14)
plt.tight_layout()
plt.show()
```

**Insight:**
<br>
Berdasarkan visualisasi lineplot di atas terdapat pola naik-turun secara berulang setiap tahun untuk SO2, NO2, CO dan 03, pola tersebut terjadi karena adanya faktor musiman seperti cuaca, aktivitas industri atau polusi kendaraan. Garis yang naik berarti polusi udara semakin buruk sedangkan garis yang turun berarti kualitas udara membaik. Konsentrasi rata-rata CO selalu lebih tinggi dari polutan gas lainnya, paling tinggi sebelum akhir tahun 2015 dan paling rendah di paruh pertengahan awal tahun 2016. Hal ini dapat menunjukkan bahwa sumber emisi CO lebih dominan atau lebih sulit dikendalikan dibandingkan dengan polutan lainnya. Puncak tertinggi sebelum akhir tahun 2015 bisa terkait dengan aktivitas yang menyebabkan emisi CO meningkat, misalnya peningkatan transportasi atau pembakaran bahan bakar fosil yang lebih intensif pada periode tersebut. Sedangkan konsentrasi rata-rata polutan gas lainnya seperti SO2, NO2, dan O3 relatif sama yaitu di bawah 200 ug/m³, ini menunjukkan bahwa polusi udara dari gas-gas ini lebih terkontrol atau tidak seberat polusi CO. Polutan ini mungkin dipengaruhi oleh faktor-faktor tertentu seperti kebijakan pengurangan emisi atau perubahan sumber emisi yang lebih efisien.

### Pertanyaan 4: Bagaimana hubungan antara tekanan udara (PRES) dengan tingkat polutan partikulat (PM2.5 dan PM10)?

```python
sns.set_theme(style="whitegrid", palette="muted")

plt.figure(figsize=(14, 8))

sns.regplot(
    data=df_cleaned, 
    x='PRES', y='PM2.5', 
    scatter_kws={'alpha': 0.5, 'color': 'steelblue'}, 
    line_kws={'color': 'darkblue', 'linewidth': 2}, 
    label='PM2.5'
)

sns.regplot(
    data=df_cleaned, 
    x='PRES', y='PM10', 
    scatter_kws={'alpha': 0.5, 'color': 'orange'}, 
    line_kws={'color': 'darkorange', 'linewidth': 2}, 
    label='PM10'
)

plt.title('Hubungan antara Tekanan Udara (PRES) dan Tingkat Polutan Partikulat (PM2.5 & PM10)', 
          fontsize=16, fontweight='bold', color='navy')
plt.xlabel('Tekanan Udara (hPa)', fontsize=12)
plt.ylabel('Konsentrasi (\u03bcg/m³)', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.legend(title='Polutan', fontsize=12, title_fontsize=13)
plt.grid(visible=True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

**Insight:**
<br>
Berdasarkan visualisasi scatter plot dengan regresi di atas garis regresi cenderung turun, berarti ada hubungan negatif (ketika tekanan udara meningkat, konsentrasi polutan menurun). Hal ini bisa terjadi karena pada kondisi tekanan udara yang lebih tinggi, fenomena atmosfer seperti pergerakan udara vertikal bisa menyebabkan polutan terangkat ke atmosfer yang lebih tinggi, sehingga konsentrasi polutan di permukaan berkurang. Tekanan udara yang lebih tinggi sering kali terjadi pada cuaca cerah dengan sedikit angin, yang memungkinkan polutan untuk mengendap atau mengurai lebih cepat, menurunkan konsentrasinya di dekat permukaan.

### Pertanyaan 5: Apakah kecepatan angin rendah (< 1 m/s) cenderung menyebabkan penumpukan polutan partikulat (PM2.5 dan PM10)?

```python
df_cleaned['wind_speed_category'] = df_cleaned['WSPM'].apply(lambda x: 'Low (<1 m/s)' if x < 1 else 'High (≥1 m/s)')

sns.set_theme(style="whitegrid", palette="muted")

plt.figure(figsize=(14, 8))

plt.subplot(1, 2, 1)
sns.boxplot(
    data=df_cleaned, 
    x='wind_speed_category', y='PM2.5', 
    hue='wind_speed_category',
    dodge=False,
    palette={'Low (<1 m/s)': 'steelblue', 'High (≥1 m/s)': 'darkorange'},  
    legend=False
)
plt.title('Distribusi PM2.5 Berdasarkan Kecepatan Angin', fontsize=14, fontweight='bold')
plt.xlabel('Kategori Kecepatan Angin', fontsize=12)
plt.ylabel('Konsentrasi PM2.5 (\u03bcg/m³)', fontsize=12)

plt.subplot(1, 2, 2)
sns.boxplot(
    data=df_cleaned, 
    x='wind_speed_category', y='PM10', 
    hue='wind_speed_category',
    dodge=False,
    palette={'Low (<1 m/s)': 'steelblue', 'High (≥1 m/s)': 'darkorange'},  
    legend=False
)
plt.title('Distribusi PM10 Berdasarkan Kecepatan Angin', fontsize=14, fontweight='bold')
plt.xlabel('Kategori Kecepatan Angin', fontsize=12)
plt.ylabel('Konsentrasi PM10 (\u03bcg/m³)', fontsize=12)
plt.tight_layout()
plt.show()
```

**Insight:**
<br>
Berdasarkan visualisasi boxplot di atas, median untuk konsentrasi rata-rata PM2.5 dan PM10 lebih tinggi di kecepatan angin rendah. Hal ini menunjukkan bahwa PM2.5 dan PM10 memiliki median yang lebih tinggi dan sebaran yang lebih luas pada kecepatan angin rendah, ini bisa berarti bahwa kecepatan angin yang rendah tidak cukup untuk menyebarkan polusi dengan efektif, sehingga konsentrasi partikel lebih terkonsentrasi. Median lebih rendah pada kecepatan angin tinggi, sehingga angin bisa berfungsi untuk menyebarkan polusi dan mengurangi konsentrasinya.

### Pertanyaan 6: Bagaimana dampak hujan deras (>20 mm) terhadap konsentrasi partikulat PM2.5 dan PM10?

```python
rain_heavy = df_cleaned[df_cleaned['RAIN'] > 20]

avg_pm_heavy = rain_heavy[['PM2.5', 'PM10']].mean()

pm_data = pd.DataFrame({
    'Konsentrasi PM2.5 dan PM10': [avg_pm_heavy['PM2.5'], avg_pm_heavy['PM10']],
},
index=['PM2.5', 'PM10'])

plt.figure(figsize=(8, 6))

sns.barplot(x=pm_data.index, y=pm_data['Konsentrasi PM2.5 dan PM10'], hue=pm_data.index, palette='Blues')

plt.title('Rata-rata Konsentrasi PM2.5 dan PM10 pada Hujan Deras (>20 mm)')
plt.ylabel('Konsentrasi (µg/m³)')
plt.tight_layout()
plt.show()
```

**Insight:**
<br>
Berdasarkan visualisasi barplot di atas, bar untuk PM10 lebih tinggi dari PM2.5 maka ini menunjukkan bahwa konsentrasi PM10 lebih tinggi dibandingkan PM2.5 selama hujan deras. Dalam kondisi hujan deras, partikel berukuran lebih besar lebih dominan di atmosfer, yang bisa disebabkan oleh sumber polusi yang menghasilkan partikel besar, seperti debu atau polusi dari kendaraan dan industri. Hujan dapat membantu mengendapkan partikel kecil seperti PM2.5, sehingga konsentrasi PM2.5 mungkin lebih rendah dibandingkan PM10 yang lebih besar dan lebih mudah terjebak di atmosfer. Meskipun PM10 lebih besar dari PM2.5, PM2.5 tetap menjadi polutan yang lebih berbahaya karena dapat mempengaruhi kesehatan dalam jangka panjang, terutama sistem pernapasan bagian atas (hidung, tenggorokan, dan saluran pernapasan besar). PM2.5 juga memiliki ukuran yang sangat kecil, yang memungkinkannya masuk ke dalam saluran pernapasan dalam dan bahkan mencapai alveoli (kantung udara di paru-paru).

### Pertanyaan 7: Apakah penurunan suhu (TEMP) di malam hari mempengaruhi kadar polutan berbahaya seperti CO, NO2 dan SO2?

```python
night_time = df_cleaned[(df_cleaned['hour'] >= 18) | (df_cleaned['hour'] <= 6)]

plt.figure(figsize=(14, 8))

plt.subplot(1, 3, 1)
sns.scatterplot(data=night_time, x='TEMP', y='CO', color='blue')
plt.title('Suhu (TEMP) vs CO')
plt.xlabel('Suhu (°C)')
plt.ylabel('Kadar CO (µg/m³)')

plt.subplot(1, 3, 2)
sns.scatterplot(data=night_time, x='TEMP', y='NO2', color='green')
plt.title('Suhu (TEMP) vs NO2')
plt.xlabel('Suhu (°C)')
plt.ylabel('Kadar NO2 (µg/m³)')

plt.subplot(1, 3, 3)
sns.scatterplot(data=night_time, x='TEMP', y='SO2', color='red')
plt.title('Suhu (TEMP) vs SO2')
plt.xlabel('Suhu (°C)')
plt.ylabel('Kadar SO2 (µg/m³)')

plt.tight_layout()
plt.show()
```

**Insight:**
<br>
Berdasarkan visualisasi scatter plot di atas, CO memiliki titik yang berada dalam posisi sejajar horizontal, SO2 di suhu 0-40 titiknya menjadi lebih acak padahal di luar rentang tersebut titiknya sejajar, sedangkan NO2 titiknya sangat acak. Titik sejajar horizontal pada grafik CO menunjukkan bahwa kadar CO cenderung stabil dan tidak terpengaruh secara signifikan oleh perubahan suhu pada malam hari. Kadar CO tidak terpengaruh oleh suhu pada malam hari, yang mengindikasikan bahwa sumber emisi CO lebih dominan dibandingkan faktor suhu. Pada grafik SO2, titik yang lebih acak pada suhu 0-40°C, namun sejajar di luar rentang tersebut, menunjukkan bahwa pada suhu rendah hingga sedang (0-40°C), kadar SO2 cenderung lebih bervariasi dan dipengaruhi oleh faktor lain selain suhu. Kadar SO2 lebih bervariasi pada suhu 0-40°C, tetapi di luar rentang tersebut, faktor lain (seperti aktivitas industri atau kondisi cuaca) lebih berperan dalam mengatur kadar SO2. Titik yang sangat acak pada grafik NO2 menunjukkan bahwa tidak ada pola atau hubungan yang jelas antara suhu dan kadar NO2. Hal ini bisa menunjukkan bahwa NO2 dipengaruhi oleh banyak faktor lain, seperti lalu lintas kendaraan, aktivitas industri, dan faktor meteorologi seperti angin atau kelembaban, yang lebih dominan dibandingkan suhu pada malam hari.

## Conclusion

- **Conclution pertanyaan 1**: Terdapat tren perubahan polutan partikulat (PM2.5 dan PM10) yang berfluktuasi selama periode 2013 hingga 2017, dengan pola naik-turun yang terjadi setiap tahunnya. Fluktuasi ini dipengaruhi oleh faktor musiman seperti cuaca, aktivitas industri, dan polusi kendaraan. PM10 memiliki konsentrasi rata-rata yang lebih tinggi dibandingkan PM2.5, yang bisa disebabkan oleh debu, angin, atau aktivitas konstruksi. Meskipun PM10 merupakan partikel yang lebih besar, keduanya sama-sama berbahaya bagi kesehatan. PM2.5 lebih berbahaya karena dapat masuk lebih dalam ke paru-paru. Pada akhir tahun 2013, 2014, dan awal tahun 2016, kualitas udara membaik dengan konsentrasi PM2.5 dan PM10 yang lebih rendah.
- **Conclution pertanyaan 2**: Tren polutan partikulat (PM2.5 dan PM10) menunjukkan perbedaan yang signifikan antara pagi hari dan sore hari di setiap musim. PM10 memiliki konsentrasi terendah di musim dingin pagi dan tertinggi di musim semi pagi. Hal ini dipengaruhi oleh curah hujan, salju, suhu rendah, dan aktivitas pertanian yang lebih banyak di musim semi. Sementara itu, PM2.5 paling rendah di musim dingin pagi karena aktivitas manusia yang terbatas dan emisi polutan yang rendah, namun paling tinggi di musim panas pagi, di mana suhu tinggi meningkatkan pembakaran bahan bakar dan aktivitas manusia, menghasilkan lebih banyak emisi dari kendaraan dan industri.
- **Conclution pertanyaan 3**: Terdapat tren perubahan polutan gas (SO2, NO2, CO, O3) yang berfluktuasi secara berulang setiap tahun antara 2013 hingga 2017, dengan pola naik-turun yang dipengaruhi oleh faktor musiman seperti cuaca, aktivitas industri, dan polusi kendaraan. Konsentrasi rata-rata CO selalu lebih tinggi dibandingkan polutan gas lainnya, dengan puncak tertinggi sebelum akhir tahun 2015. Hal ini menunjukkan bahwa emisi CO lebih dominan atau sulit dikendalikan, mungkin terkait dengan peningkatan transportasi atau pembakaran bahan bakar fosil. Sementara itu, SO2, NO2, dan O3 memiliki konsentrasi rata-rata yang lebih rendah (di bawah 200 µg/m³), yang mengindikasikan bahwa polusi dari gas-gas ini lebih terkontrol atau lebih sedikit dipengaruhi oleh sumber emisi tertentu.
- **Conclution pertanyaan 4**: Terdapat hubungan negatif antara tekanan udara (PRES) dan tingkat polutan partikulat (PM2.5 dan PM10), yang ditunjukkan oleh garis regresi yang cenderung turun. Ketika tekanan udara meningkat, konsentrasi polutan menurun. Hal ini disebabkan oleh fenomena atmosfer seperti pergerakan udara vertikal yang mengangkat polutan ke atmosfer yang lebih tinggi, mengurangi konsentrasi polutan di permukaan. Tekanan udara yang lebih tinggi sering kali terjadi pada cuaca cerah dengan sedikit angin, yang memungkinkan polutan untuk mengendap atau terurai lebih cepat, sehingga konsentrasi polutan di permukaan berkurang.
- **Conclution pertanyaan 5**: Kecepatan angin rendah (< 1 m/s) cenderung menyebabkan penumpukan polutan partikulat (PM2.5 dan PM10), seperti yang terlihat pada visualisasi boxplot, di mana median konsentrasi PM2.5 dan PM10 lebih tinggi pada kecepatan angin rendah. Hal ini mengindikasikan bahwa angin dengan kecepatan rendah tidak cukup efektif untuk menyebarkan polusi, sehingga konsentrasi partikel lebih terkonsentrasi di area tersebut. Sebaliknya, pada kecepatan angin tinggi, median konsentrasi polutan lebih rendah, menunjukkan bahwa angin yang lebih kuat dapat menyebarkan polusi dan mengurangi konsentrasi partikel di udara.
- **Conclution pertanyaan 6**: Selama hujan deras (>20 mm), konsentrasi partikulat PM10 lebih tinggi dibandingkan PM2.5, seperti yang ditunjukkan dalam visualisasi barplot. Hal ini disebabkan oleh dominasi partikel berukuran lebih besar, seperti debu dan polusi dari kendaraan atau industri, yang lebih mudah terjebak di atmosfer. Hujan dapat membantu mengendapkan partikel kecil seperti PM2.5, sehingga konsentrasinya lebih rendah. Meskipun PM10 lebih besar, PM2.5 tetap menjadi polutan yang lebih berbahaya karena dapat memasuki saluran pernapasan dalam dan berdampak buruk pada kesehatan jangka panjang, terutama pada sistem pernapasan bagian dalam (seperti alveoli di paru-paru).
- **Conclution pertanyaan 7**: Penurunan suhu (TEMP) di malam hari tidak memiliki pengaruh signifikan terhadap kadar polutan berbahaya seperti CO, NO2, dan SO2. Grafik menunjukkan bahwa kadar CO cenderung stabil tanpa dipengaruhi suhu malam hari, mengindikasikan bahwa sumber emisi CO lebih dominan. Kadar SO2 bervariasi pada suhu 0-40°C, tetapi di luar rentang tersebut, faktor lain seperti aktivitas industri atau kondisi cuaca lebih mempengaruhi kadar SO2. Sementara itu, grafik untuk NO2 menunjukkan ketidakteraturan, yang mengindikasikan bahwa NO2 dipengaruhi oleh berbagai faktor lain seperti lalu lintas kendaraan dan aktivitas industri, bukan suhu malam hari.
